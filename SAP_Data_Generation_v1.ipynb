# Code to Generate Orderdate using Monthly Distribution

```python
# Function to generate a random date between two dates
def random_date(start_date, end_date):
    delta = end_date - start_date
    random_days = random.randint(0, delta.days)
    return start_date + timedelta(days=random_days)
```

```python
import random
from datetime import datetime, timedelta

def generate_order_dates(year, no_of_orders, month_distribution, exclude_months):
    # Filter out the excluded months from the distribution
    filtered_distribution = {month: prob for month, prob in month_distribution.items() if month not in exclude_months}
    
    # Normalize the filtered distribution
    total_prob = sum(filtered_distribution.values())
    normalized_distribution = {month: prob / total_prob for month, prob in filtered_distribution.items()}
    
    # Generate cumulative distribution for the filtered months
    cumulative_distribution = []
    cumulative_sum = 0
    for month in range(1, 13):
        if month in normalized_distribution:
            cumulative_sum += normalized_distribution[month]
            cumulative_distribution.append((month, cumulative_sum))
    
    # Function to get a random month based on the distribution
    def get_random_month():
        rand = random.random()
        for month, cum_prob in cumulative_distribution:
            if rand <= cum_prob:
                return month
    
    # Generate random dates
    random_dates = []
    for _ in range(no_of_orders):
        random_month = get_random_month()
        # Get a random day in the selected month
        while True:
            random_day = random.randint(1, 31)
            try:
                random_date = datetime(year=year, month=random_month, day=random_day)
                random_dates.append(random_date.strftime("%m-%d-%Y"))
                break
            except ValueError:
                continue
    
    return random_dates
```

```python
# Example usage
year = 2022
no_of_orders = 22950
month_distribution = {
    1: 0.11, 2: 0.06, 3: 0.09, 4: 0.06, 5: 0.12, 6: 0.06, 7: 0.08, 8: 0.07, 9: 0.13, 10: 0.06, 11: 0.09, 12: 0.07
}
exclude_months = []  # Example months to exclude

orderdate_2022 = generate_order_dates(year, no_of_orders, month_distribution, exclude_months)
```

```python
def add_days(input_date, range1, range2):

    if isinstance(input_date, str):    
    # Convert orderdate string back to datetime object
        input_date = datetime.strptime(input_date, '%m-%d-%Y')

    # Generate a random number of days 
    additional_days = random.randint(range1, range2)

    # Calculate the new date
    new_date = input_date + timedelta(days=additional_days)

    # Convert the new date back to string if needed
    new_date = new_date.strftime('%m-%d-%Y')

    return new_date
```

```python
# Function to determine Payment_Terms based on CreditLimit
def determine_payment_range(rating):
    if rating < 50000:
        return 15
    elif 50000 <= rating < 75000:
        return 25
    elif 75000 <= rating < 90000:
        return 30
    else:
        return 45
```

```python
Supplier_Delivery = {'Sure Source' : [5,10]
,'Prime Partners' : [3,7]
,'Wholesale Wizards' : [1,3]
,'Supply Solutions' : [5,15]
,'Procure Plus' : [10,15]
}
```

```python
# List to store orders
orders_fulfill = []
items_fulfill = []
shipment_fulfill = []
invoice_fulfill = []

# Generate orders
for i in range(85000):
    # Filter products matching the customer's type
    customer = df_customers.sample(n=1)
    matching_products = df_product[df_product['PRODUCTCATEGORY'] == customer.iloc[0]['CUSTOMERTYPE']]
    num_items = random.choice([1,2,3])  # Each order has a random number of items between 1 and 3
    
    supplier_name = matching_products.iloc[0]['SUPPLIERNAME']
    supplier_delivery_startrange, supplier_delivery_endrange = Supplier_Delivery[supplier_name]
    
    customerid = customer.iloc[0]['CUSTOMERID']
    orderid = 'OR-' + fake.uuid4()
    #orderdate = orders[i]
    orderdate = orderdate[i]
    #orderdate = random_date(start_date, end_date).strftime('%m-%d-%Y')
    orderstatus = 'Approved'
    orderamount = 0
    ordercount = 0
    invoice_range = 0
    
    shipmentid = 'SP-' + fake.uuid4()
    shipmentdate = add_days(orderdate, supplier_delivery_startrange, supplier_delivery_endrange)

    delivery_range = 5
    if supplier_name in ['Prime Partners','Wholesale Wizards']:
        delivery_range = 2
        
    deliverydate = add_days(shipmentdate, 2 ,3)
    deliveredOn = add_days(deliverydate,1,delivery_range) 

    invoiceid = 'IN-' + fake.uuid4()
    invoicedate = add_days(deliveredOn,0,1)

    credit_limit = customer.iloc[0]['CREDITLIMIT']
    #print (customerid, credit_limit)
    invoice_range = determine_payment_range(credit_limit)
    invoiceduedate = add_days(invoicedate,invoice_range-1, invoice_range)
        
    for _ in range(num_items):
        products = matching_products.sample(n=1)
        product_type = products.iloc[0]['PRODUCTTYPE']
        quantity = random.randint(1, products.iloc[0]['QUANTITY'])  # Generate a random quantity for each order item
        unit_price = products.iloc[0]['UNITPRICE']
        total_price = quantity * unit_price
        orderamount += total_price
        ordercount += 1
        matching_products = matching_products[matching_products['PRODUCTTYPE'] != product_type]
        orderitemid = 'OI-' + fake.uuid4()


        items_dict = {
            'ORDERID' : orderid,
            'ORDERITEMID': orderitemid,
            'PRODUCTID': products.iloc[0]['PRODUCTID'],
            'PRODUCTNAME' : products.iloc[0]['PRODUCTTYPE'],
            'ORDERQUANTITY' :quantity,
            'UNITPRICE' : unit_price,
            'ORDERVALUE' : total_price
        }

        items_fulfill.append(items_dict)
    
    payment_status = 'ONTIME_PAYMENT'
    if orderamount > 0.95 * credit_limit:  # If order amount exceeds 95% of the credit limit
        paymentdate = add_days(invoicedate,invoice_range+10, invoice_range+15)  # Delay payment by 10 days
        payment_status = 'LATE_PAYMENT'
    elif orderamount > 0.90 * credit_limit:  # If order amount exceeds 90% of the credit limit
        paymentdate = add_days(invoicedate,invoice_range+5, invoice_range+10)  # Delay payment by 15 days
        payment_status = 'LATE_PAYMENT'
    else:
        paymentdate = add_days(invoicedate,1, invoice_range) #No delay in payments
    
        

    orders_dict = {
        'ORDERID' : orderid,
        'CUSTOMERID': customerid,
        'CUSTOMERNAME': customer.iloc[0]['CUSTOMERNAME'],
        'PRODUCTID': matching_products.iloc[0]['PRODUCTID'],
        'PRODUCTCATEGORY': matching_products.iloc[0]['PRODUCTCATEGORY'],
        'ORDERDATE': orderdate,
        'ORDERSTATUS': orderstatus,
        'ORDERAMOUNT' :orderamount,
        'ORDERCOUNT' : ordercount
    }

    orders_fulfill.append(orders_dict)

    shipment_dict = {
        'SHIPMENTID': shipmentid,
        'ORDERID' : orderid,
        'SHIPMENTDATE': shipmentdate,
        'DELIVERYDATE': deliverydate,
        'DELIVEREDON' :deliveredOn,
        'CARRIER' : random.choice(['FedEx', 'UPS', 'DHL', 'USPS']),
        'SHIPMENTSTATUS' : 'Delivered'
    }

    shipment_fulfill.append(shipment_dict)

    invoice_dict = {
        'INVOICEID': invoiceid,
        'ORDERID' : orderid,
        'INVOICEDATE': invoicedate,
        'INVOICEDUEDATE': invoiceduedate,
        'PAYMENTDATE' : paymentdate,
        'INVOICESTATUS': payment_status,
        'PAYMENTMETHOD': random.choice(['Wire Transfers','Cards','Checks','Mobile Payments'])
    }

    invoice_fulfill.append(invoice_dict)
```
